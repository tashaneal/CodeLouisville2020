```python
import pandas as pd
import sqlite3
import os
os.getcwd()

#create the sqlite DB
db = sqlite3.connect ("MyDemo.db")

#read CSV into dataframe
df1 = pd.read_csv ("titanic.csv", header=None)

#convert the pandas dataframe to a sqlite table
df1.to_sql("Survived",db, if_exists="replace")

def run_query (query):
    return pd.read_sql_query(query,db)

os.getcwd()
```

```python
query = """
SELECT *
FROM sqlite_master
ORDER BY name;
"""
run_query(query)
```
